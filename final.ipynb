```python
import pandas as pd

df_konnagar = pd.read_csv("konnagar.csv")
df_konnagar['Location'] = 'Konnagar'
df_liluah = pd.read_csv("liluah.csv")
df_liluah['Location'] = 'Liluah'
df_shibpur = pd.read_csv("shibpur.csv")
df_shibpur['Location'] = 'Shibpur'

df = pd.concat([df_konnagar, df_liluah, df_shibpur], axis=0)
```

```python
df.sample(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.swarmplot(data=df,x='Humidit',y='Air_Quality(PPM)',hue= 'Location')
plt.xlabel('Humidity')
plt.ylabel('Air Quality')
plt.title('Air Quality vs Humidity graph')
plt.show()
```
